```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


try:
    dataset = pd.read_csv('TUANDROMD.csv')
    print("Dataset berhasil diimpor.")
    print("\nLima baris pertama dataset:")
    print(dataset.head())
    print("\nInformasi tipe data kolom:")
    print(dataset.info())
except FileNotFoundError:
    print("ERROR: Pastikan file 'TUANDROMD.csv' berada dalam direktori yang sama.")
    exit()


dataset.dropna(subset=['Label'], inplace=True)

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print("\nShape Matriks X (Fitur):", X.shape)
print("Shape Vektor y (Target):", y.shape)


print("\nJumlah Missing Values per Kolom Setelah Menghapus Baris dengan Label Kosong:")
print(dataset.isnull().sum().tail())



le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("\nLabel Kategori Asli:", np.unique(y))
print("Label Kategori Hasil Encoding (y_encoded):", np.unique(y_encoded))


X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

print("\n--- Hasil Pembagian Data ---")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)


sc = StandardScaler()


X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)

print("\n--- Hasil Feature Scaling (StandardScaler) ---")

print("X_train sebelum scaling (Contoh Baris 1, 5 kolom):", X_train[0, :5])
print("X_train setelah scaling (Contoh Baris 1, 5 kolom):", X_train_scaled[0, :5])
```
